# CHECKING NORMAL DISTRIBUTION USING QQ PLOT

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
sns.kdeplot(df['sepal_length'])
```

```python
temp = sorted(df['sepal_length'].tolist())
```

```python
y_quant = []
for i in range(1,101):
  y_quant.append(np.percentile(temp,i))
```

```python
sample = np.random.normal(loc = 0, scale = 1, size = 1000)
```

```python
x_quant = []
for i in range(1,101):
  x_quant.append(np.percentile(sample,i))
```

```python
sns.scatterplot(x = x_quant, y = y_quant)
```

```python
# using inbuilt function

import statsmodels.api as sm
import matplotlib.pyplot as plt

fig = sm.qqplot(df['sepal_length'], line = '45', fit = True)

plt.show()
```

# CHECKING UNIFORM DISTRIBUTION USING QQ PLOT

```python
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(low = 0, high = 1, size = 1000)
```

```python
plt.hist(x)
```

```python
params = stats.uniform.fit(x)
dist = stats.uniform(loc = params[0], scale = params[1])
```

```python
fig = sm.qqplot(x , line = '45', dist = dist)
```

# CHECKING PARETO DISTRIBUTION USING LOG LOG PLOT

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the parameters of the Pareto distribution

alpha = 3;
xm =1;
```

```python
x = np.linspace(0.1 , 10, 1000)
y = alpha * (xm**alpha)/(x**(alpha + 1))
```

```python
plt.plot(x,y)
```

```python
# checking time
```

```python
plt.plot(np.log(x), np.log(y))
```

# CHECKING PARETO DISTRIBUTION USING Q Q PLOT

```python
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
alpha = 2
xm = 1
```

```python
x = stats.pareto.rvs(b = alpha, scale = xm, size = 1000)
```

```python
plt.hist(x)
```

```python
params = stats.pareto.fit(x , floc =0)
dist = stats.pareto(b=params[0], scale=params[2])
```

```python
fig = sm.qqplot(x, dist=dist, line='45')
```
